# Parsing

```python
def parse_cl(cl_file):
    with open(cl_file,"r",encoding="utf-8-sig") as f:
        cl_string = f.read()
    try:
        r_cl_string = "["+cl_string[:-2]+"]"
        cl_list = json.loads(r_cl_string)
    except:
        cl_string = cl_string.replace("\n",",\n")
        cl_string = "["+cl_string[:-2]+"]"
        cl_list = json.loads(cl_string)
    return cl_list
```

```python
def list_to_dict(cl_list):
    control_log_dict = {}
    for i in range(int(len(cl_list)/2)):
        this_frame = cl_list[i*2]
        this_action = cl_list[(i*2)+1]
        this_frame_number = this_frame["frame_number"]
        control_log_dict[this_frame_number]={"frame":this_frame,"action":this_action}
    return control_log_dict
```

```python
def feedback_info(control_log_dict):
    up_signals = []
    up_attempts = []
    down_signals = []
    down_attempts = []
    highlight_destroy = None
    ask_before = None
    ask_after = None
    
    for frame in list(control_log_dict.keys()):
        this_frame = control_log_dict[frame]["frame"]
        if this_frame['tried_signal_up']==True:
            up_attempts.append(frame)
        if this_frame['tried_signal_down']==True:
            down_attempts.append(frame)
        if this_frame['signal_up']==True:
            up_signals.append(frame)
        if this_frame['signal_down']==True:
            down_signals.append(frame)
        if ask_after == None:
            highlight,before,after = action_from_game_state(this_frame)
        if highlight_destroy == None and highlight:
            highlight_destroy = frame
        if ask_before == None and before:
            ask_before = frame
        if ask_after == None and after:
            ask_after = frame
        
    
    frames_before_to_up = None
    frames_before_to_down = None
    frames_after_to_up = None
    frames_after_to_down = None
    for up in up_signals:
        if up > ask_before and frames_before_to_up == None:
            frames_before_to_up = up - ask_before
        if up > ask_after and frames_after_to_up == None:
            frames_after_to_up = up - ask_after
    for down in down_signals:
        if down > ask_before and frames_before_to_down == None:
            frames_before_to_down = down - ask_before
        if down > ask_after and frames_after_to_down == None:
            frames_after_to_down = down - ask_after
    data = {'up_signals_list':up_signals,'up_signals_count':len(up_signals),
            'up_attempts_list':up_attempts,'up_attempts_count':len(up_attempts),
            'down_signals_list':down_signals,'down_signals_count':len(down_signals),
            'down_attempts_list':down_attempts,'down_attempts_count':len(up_attempts),
            'highlight_destroy':highlight_destroy,'ask_before':ask_before,'ask_after':ask_after,
            'frames_before_to_up':frames_before_to_up,'frames_before_to_down':frames_before_to_down,
            'frames_after_to_up':frames_after_to_up,'frames_after_to_down':frames_after_to_down
           }
    return data
```

```python
GAME_ENEMIES = 9*4*3*2 # across * row per color * color * sides
MIDDLE = 600 # middle of x axis

def action_from_game_state(game_state):
        enemies_left_positions = game_state['enemies_left_positions']
        enemies_right_positions = game_state['enemies_right_positions']
        num_left_enemies = len(enemies_left_positions)
        num_right_enemies = len(enemies_right_positions)

        total_enemies = num_left_enemies+num_right_enemies

        if total_enemies < GAME_ENEMIES*.75 and game_state['ai_position']<MIDDLE*.75:
            highlight = True
        else:
            highlight = False

        # asking for feedback before crossing to help
        if total_enemies < GAME_ENEMIES*.875:
            before = True
        else:
            before = False
        # after for feedback after crossing to help
        if total_enemies <= GAME_ENEMIES*.65 and game_state["ai_position"] > MIDDLE*1.1:
            after = True
        else:
            after = False
        return highlight,before,after
```
